```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
Q1 = df['OS'].reset_index()
sns.countplot(x='OS', data=Q1)
plt.xlabel('Name Of Operating System')
plt.ylabel('Count Of Computers')
plt.title('Count Of Operatings Systems Computers')
plt.show()
```

```python
Q2 = df['Company'].reset_index()
sns.countplot(x='Company', data=Q2)
plt.xlabel('Name Of Company')
plt.ylabel('Count Of Computer')
plt.title('Count Of Companies Computers')
plt.show()
```

```python
Q2 = df['IPSpanel'].value_counts()
plt.pie(Q2, labels=Q2.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of IPSpanel')
plt.show()
```

```python
Q4=df['Inches'].value_counts().reset_index()
sns.histplot(x='Inches' , data=Q4, palette='viridis')
plt.xlabel('Screen Size [Inche]')
plt.ylabel('Count Of Computers')
plt.title('Screen Size and Count of Computers')
plt.show()
```

```python
Q5=df.groupby('Company')['Price_euros'].mean().head(7).reset_index()
sns.barplot(x='Company' , y='Price_euros' , data=Q5)
plt.xlabel('Company Name')
plt.ylabel('Computers Avg Price [Euros]')
plt.title('Avg Price of Companies Computers [TOP 7]')
plt.show()
```

```python
Q6=df.groupby('RetinaDisplay')['Price_euros'].mean().reset_index()
sns.barplot(x='RetinaDisplay' , y='Price_euros' , data=Q6)
plt.xlabel('RetinaDisplay [Yes | NO]')
plt.ylabel('Computers Avg Price [Euros]')
plt.title('Retina Display Computers Avg Price')
plt.show()
```

```python
Q7=df.groupby('Ram')['Price_euros'].mean().reset_index()
sns.lineplot(x='Ram' , y='Price_euros' , data=Q7)
plt.xlabel('Ram of computers')
plt.ylabel('Computers Avg Price [Euros]')
plt.title('Computers Avg Price [Euros] Based On RAM')
plt.show()
```

```python
Q8=df.groupby('GPU_company')['Price_euros'].mean().reset_index()
sns.lineplot(x='GPU_company' , y='Price_euros' , data=Q8)
plt.xlabel('GPU company Name')
plt.ylabel('Computers Avg Price [Euros]')
plt.title('Computers Avg Price [Euros] Based On GPU')
plt.show()
```

```python
Q9 = df['Touchscreen'].value_counts().reset_index()
Q5.columns = ['Touchscreen', 'Count']
sns.histplot(x='Touchscreen', weights='Count', data=Q5 , palette='viridis')

plt.xlabel('Screen Type [Touchscreen]')
plt.ylabel('Count of Computers')
plt.title('Count of Computers Based On Screen Type')
plt.show()
```

```python
Q10=df.groupby('CPU_company')['CPU_freq'].mean().reset_index()
sns.lineplot(x='CPU_company', y='CPU_freq',data=Q10)
plt.xlabel('CPU Companies')
plt.ylabel('AVG CPU Frequency ')
plt.show()
```

```python
Q11 = df['GPU_company'].value_counts().reset_index()
Q11.columns = ['GPU_company', 'Count']
sns.barplot(x='GPU_company', y='Count', data=Q11)
plt.xlabel('GPU Company')
plt.ylabel('Count of Computers')
plt.title('Distribution of Laptops by GPU Company')
plt.show()
```

```python
Q12=df.groupby('Ram')['Price_euros'].mean().reset_index()
sns.scatterplot(x='Ram' , y='Price_euros' , data=Q12)
plt.xlabel('Ram Of The Computers')
plt.ylabel('Price in Euros')
plt.title('Prize Distribution Based On RAM')
```

```python
Q13=df.groupby('OS')['Weight'].mean().reset_index()
sns.barplot(x='OS', y='Weight', data=Q13)
plt.xlabel('Product')
plt.ylabel('Weight')
plt.title('Weight Distribution Across Products')
plt.show()
```

```python
Q14=df['Price_euros'].value_counts().reset_index()
sns.histplot(x='Price_euros' , data=Q14, palette='viridis' ,bins=50)
plt.xlabel('Prize Of ')
plt.ylabel('Count Of Computers')
plt.show()
```

```python
Q15=df['PrimaryStorage'].value_counts().reset_index()
sns.histplot(x='PrimaryStorage' , data=Q15)
plt.xlabel('Primary Storage Computer')
plt.ylabel('Count Of Computers')
plt.show()
```

```python
Q15=df.groupby('Company')['CPU_freq'].mean().reset_index()
sns.boxplot(x='Company', y='CPU_freq', data=Q15)
plt.xlabel('Company')
plt.ylabel('CPU Frequency (GHz)')
plt.title('CPU Frequency Distribution Across Companies')
plt.show()
```

```python
Q16=df.groupby('Ram')['CPU_freq'].mean().reset_index()
sns.boxplot(x='Ram', y='CPU_freq', data=Q16)
plt.xlabel( ' Ram Company')
plt.ylabel('CPU Frequency (GHz)')
plt.title('CPU Frequency Distribution Across RAM')
plt.show()
```

```python
Q17 = df.groupby('Product')['Price_euros'].mean().head(7).reset_index()
sns.lineplot(x='Product', y='Price_euros', data=Q17)
plt.xlabel('Product')
plt.ylabel('AVG Price In Euros')
plt.title('Avg Price Of Product Top 7')
plt.xticks(rotation=45, ha='right')
plt.show()

Q17B = df.groupby('Product')['Price_euros'].mean().tail(7).reset_index()
sns.lineplot(x='Product', y='Price_euros', data=Q17B)
plt.xlabel('Product')
plt.ylabel('AVG Price In Euros')
plt.title('Avg Price Of Product Tail 7')
plt.xticks(rotation=45, ha='right')
plt.show()
```
